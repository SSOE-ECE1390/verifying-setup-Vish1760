```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt

#Grayscale
apollo_image = cv2.imread('/Users/vishbish/github-classroom/SSOE-ECE1390/verifying-setup-Vish1760/Data/Apollo_11_Launch.jpg', 0)
apollo_image_corrected = cv2.cvtColor(apollo_image, cv2.COLOR_BGR2RGB)
plt.imshow(apollo_image_corrected)
plt.title("Apollo Launch")
```

```python
#Cropping
apollo_cropped  = apollo_image_corrected [0:800, 400:800]
plt.imshow(apollo_cropped)
```

```python
#Brightness
matrix = np.ones(apollo_cropped.shape, dtype="uint8") * 80
apollo_brighter = cv2.add(apollo_cropped, matrix)
plt.imshow(apollo_brighter);plt.title("Brighter")
```

```python
#Rotated
apollo_rotated = cv2.flip(apollo_brighter, 0)
plt.imshow(apollo_rotated);plt.title("Rotated")
```

```python
#Saved
cv2.imwrite("apollo_rotated.png", apollo_rotated)
os.remove("apollo_rotated.png")
```
